```python
import pandas as pd
import numpy as np
```

```python
topics = pd.read_csv('topics.csv')
topics.head()
```

```python
merged_topics = topics.groupby('topic_id')['text'].apply(lambda x: ' '.join(x)).reset_index()
merged_topics.head()
```

```python
# save into a new csv file
merged_topics.to_csv('topics.csv', index=False)
```
